## Abstract Statistics

```python
t3 = pd.read_csv(config.TABLES_DIR / 'table3_main_performance.csv')
t1 = pd.read_csv(config.TABLES_DIR / 'table1_dataset_summary.csv')
t8 = pd.read_csv(config.TABLES_DIR / 'table8_robustness.csv')

best = t3.loc[t3['pr_auc_mean'].idxmax()]
baseline_rob = t8[t8['setting'] == 'Baseline'].iloc[0]
narrow_rob   = t8[t8['setting'] == 'Narrow Barriers'].iloc[0]

print('=' * 60)
print('ABSTRACT STATISTICS')
print('=' * 60)
print(t1.set_index('Item')['Value'].to_string())
print()
print(f'Best model      : {best["model"]}')
print(f'  PR-AUC        : {best["pr_auc_mean"]:.3f} ± {best["pr_auc_std"]:.3f}')
print(f'  ROC-AUC       : {best["roc_auc_mean"]:.3f} ± {best["roc_auc_std"]:.3f}')
print(f'  Recall        : {best["recall_mean"]:.3f} ± {best["recall_std"]:.3f}')
print(f'  MCC           : {best["mcc_mean"]:.3f} ± {best["mcc_std"]:.3f}')
print()
print(f'Robustness best : {narrow_rob["setting"]} → PR-AUC={narrow_rob["pr_auc"]:.3f}')
print('=' * 60)
```
